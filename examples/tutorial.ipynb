## Report Generation

The same analyzer can generate HTML, PDF, JSON, and CSV outputs. The HTML and PDF paths below write files into `examples/generated_reports/`.

```python
from pathlib import Path
import pandas as pd
from driftguard import DriftAnalyzer

ROOT = Path.cwd()
if not (ROOT / 'driftguard').exists() and (ROOT.parent / 'driftguard').exists():
    ROOT = ROOT.parent
DATA_DIR = ROOT / 'examples' / 'sample_data'
REPORT_DIR = ROOT / 'examples' / 'generated_reports'
REPORT_DIR.mkdir(parents=True, exist_ok=True)

reference_df = pd.read_csv(DATA_DIR / 'reference.csv', parse_dates=['signup_date'])
current_df = pd.read_csv(DATA_DIR / 'current.csv', parse_dates=['signup_date'])
analyzer = DriftAnalyzer(reference_df, current_df)
report = analyzer.run()

html_path = REPORT_DIR / 'driftguard_report.html'
pdf_path = REPORT_DIR / 'driftguard_report.pdf'
json_path = REPORT_DIR / 'driftguard_report.json'
csv_path = REPORT_DIR / 'driftguard_report.csv'

analyzer.run(report_mode='html', output_path=html_path)
analyzer.run(report_mode='pdf', output_path=pdf_path)
report.to_json(json_path)
report.to_csv(csv_path)

[html_path, pdf_path, json_path, csv_path]
```

## Interactive Window

Run this cell when you want the report to open in a standalone desktop window. If PyQt5 is not installed, DriftGuard falls back to opening the generated HTML in your browser.

```python
from pathlib import Path
import pandas as pd
from driftguard import DriftAnalyzer

ROOT = Path.cwd()
if not (ROOT / 'driftguard').exists() and (ROOT.parent / 'driftguard').exists():
    ROOT = ROOT.parent
DATA_DIR = ROOT / 'examples' / 'sample_data'
REPORT_DIR = ROOT / 'examples' / 'generated_reports'
REPORT_DIR.mkdir(parents=True, exist_ok=True)

reference_df = pd.read_csv(DATA_DIR / 'reference.csv', parse_dates=['signup_date'])
current_df = pd.read_csv(DATA_DIR / 'current.csv', parse_dates=['signup_date'])
analyzer = DriftAnalyzer(reference_df, current_df)
window_path = REPORT_DIR / 'driftguard_window_report.html'
analyzer.run(report_mode='window', output_path=window_path)
window_path
```
